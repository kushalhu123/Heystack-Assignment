```python
# NOW I AM CLICKING ON GET ALL REVIEWS
def getallreviews():
    sear=driver.find_element_by_xpath("//*[@id='customer-reviews']/div[3]/div/div/div/a")
    sear.click()
getallreviews()
#//*[@id="customer-reviews"]/div[3]/div/div/div/a
```

# HERE I AM FINDING ALL THE PATHS IN WHICH I CAN GET THE INFORMATION

```python
#FOR DATES
alldates=driver.find_elements_by_xpath("//div[@class='review-date']")
#FOR PARAGRAPH
allparas=driver.find_elements_by_xpath("//div[@class='review-text']")
#FOR RATING
allrating=driver.find_elements_by_xpath("//span[@class='average-rating']")
# FOR TITLES
alltitles=driver.find_elements_by_xpath("//div[@class='review-heading']")
# FOR NAMES
allnames=driver.find_elements_by_xpath("//span[@class='review-footer-userNickname']")
```

```python
#I AM STORING ALL THE DATA IN THE FORM OF LIST
finallist=[]
count=0
reviewcount=5
rcount=1
while True:
    try:
        newlist=[]
        newDate=alldates[count]
        newpara=allparas[count]
        newrate=allrating[reviewcount]
        newtitle=alltitles[count]
        newnames=allnames[count]
        s=newDate.text
        realdate=s.split('\n')
        #print(date[0])
        date=realdate[0]
        #print(date)
        getthedate=date.split()
        #print(getthedate)
        month=getthedate[0]
        year=getthedate[-1]
        
        #IF THE YEAR GREATET THAN 2020
        year=int(year)
        month=str(month)
        #print(year)
        #print(month)
        if year>2020:
            gothedate=realdate[0]
            newlist.append(rcount)
            newlist.append(gothedate)
            gottheparagraph=newpara.text
            newlist.append(gottheparagraph)
            gottherating=newrate.text.split('\n')[1]
            newlist.append(gottherating)
            gotthetitle=newtitle.text.split('\n')
            stitle=gotthetitle[-1]
            if str(stitle)=='stars':
                stitle='NO TITLE BY USER'
            newlist.append(stitle)
            gotthename=newnames.text
            newlist.append(gotthename)
            #NOW ADDED ALL THE CREDENTIALS
            finallist.append(newlist)
            rcount=rcount+1
            count=count+1
            reviewcount=reviewcount+1
                    
        elif (year==2020 and month=="December"):
            gothedate=realdate[0]
            newlist.append(rcount)
            newlist.append(gothedate)
            gottheparagraph=newpara.text
            newlist.append(gottheparagraph)
            gottherating=newrate.text.split('\n')[1]
            newlist.append(gottherating)
            gotthetitle=newtitle.text.split('\n')
            stitle=gotthetitle[-1]
            if str(stitle)=='stars':
                stitle='NO TITLE BY USER'
            newlist.append(stitle)
            gotthename=newnames.text
            newlist.append(gotthename)
            #NOW ADDED ALL THE CREDENTIALS
            finallist.append(newlist)
            rcount=rcount+1
            count=count+1
            reviewcount=reviewcount+1
        else:
            #rcount=rcount+1
            count=count+1
            reviewcount=reviewcount+1
            continue
    except IndexError:
        break 
```
